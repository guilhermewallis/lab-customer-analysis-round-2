```python
import pandas as pd

data = pd.read_csv('csv_files/marketing_customer_analysis.csv')
```

```python
#show shape, types, columns
print(data.shape)
print(data.dtypes)
print(data.columns)
```

```python
#column name standartization
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ', '_')

print(data.columns)
```

```python
#drop duplicated index column
print(data['unnamed:_0'])
data = data.drop('unnamed:_0', axis = 1)
```

```python
#check for nan
print(data.isna().sum())

#drop column with ~50% nan values
data = data.drop('vehicle_type', axis = 1)

#drop rows with nan values in categorical columns
data = data.dropna(subset = 'state')
data = data.dropna(subset = 'vehicle_class')

print(data.shape)
print(data.columns)
```

```python
#fill nan values in numerical columns with the column's median value
#convert from float to int
data['number_of_open_complaints'] = data['number_of_open_complaints'].fillna(data['number_of_open_complaints'].median())
data['number_of_open_complaints'] = data['number_of_open_complaints'].astype('int64')

data['months_since_last_claim'] = data['months_since_last_claim'].fillna(data['months_since_last_claim'].median())
data['months_since_last_claim'] = data['months_since_last_claim'].astype('int64')

print(data.isna().sum())
```

```python
#convert date column to datetime format
data['effective_to_date'] = pd.to_datetime(data['effective_to_date'])

#extract the month value from the date to a new column
data['month'] = data['effective_to_date'].dt.month

print(data.dtypes)
print(data.columns)
```

```python
#filter the month column by the first 3 months and assign it to new df
data_q1 = data[data['month'] <= 3]
data_q1
```

```python
# function to filter data by Q1, passing in the dataframe and the date column
def get_q1(df, col_date):
    df[col_date] = pd.to_datetime(df[col_date])    
    df['month'] = df[col_date].dt.month
    return df[df['month'] <= 3]

get_q1(data, 'effective_to_date')
```
